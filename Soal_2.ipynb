```python
import cv2
import matplotlib.pyplot as plt
from pathlib import Path

img_path = Path('Image.jpg')
if not img_path.is_file():
    print("[-] Error: File 'Image.jpg' tidak ditemukan.")
else:
    img = cv2.imread(str(img_path))
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    sizes = [512, 256, 128, 64]
    
    plt.figure(figsize=(15, 6))
    plt.suptitle("Analisis Sampling Citra: Image.jpg", fontsize=18, fontweight='bold', color='navy')

    for i, size in enumerate(sizes):
        resized = cv2.resize(img_rgb, (size, size), interpolation=cv2.INTER_AREA)
        ax = plt.subplot(1, 4, i + 1)
        ax.imshow(resized)
        ax.set_title(f"Resolusi: {size}x{size}", fontsize=12, pad=10, bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'))
        ax.axis('off')

    plt.tight_layout()
    plt.show()
```
